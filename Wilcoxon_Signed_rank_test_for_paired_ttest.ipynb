```python
import pandas as pd
import pingouin
```

```python
sample_dem_data = pd.read_feather('/content/dem_votes_potus_12_16.feather')
print(sample_dem_data.head())
```

Hypothesis: the difference between the proportion of county-level votes for the Democratic candidate in 2012 and 2016 to identify if the difference is significant

```python
# A paired t-test on dem_percent_8 and dem_percent_12
pingouin.ttest(x=sample_dem_data['dem_percent_12'],y=sample_dem_data['dem_percent_16'],paired=True,alternative='two-sided')
```

```python
# Conduct a Wilcoxon test on dem_percent_12 and dem_percent_16
wilcoxon_test_results = pingouin.wilcoxon(x=sample_dem_data['dem_percent_12'],y=sample_dem_data['dem_percent_16'],alternative='two-sided')
wilcoxon_test_results
```
